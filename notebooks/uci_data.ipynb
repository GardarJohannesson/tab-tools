## `Covertype` dataset

```python
data = fetch_ucirepo("Covertype")
```

```python
# Basic info on the features.
data.data.features.info()
```

```python
# The variable types.
data.variables
```

```python
# Each `Soil_Type` is a binary variable.
data.data.features.Soil_Type1.value_counts()
```

```python
# Only one `Soil_Type` can be true at a time (or none); in short, there are 41 soil types.
data.data.features[[f"Soil_Type{i}" for i in range(1, 40)]].sum(axis=1).value_counts()
```

```python
# The target is classification.
data.data.targets["Cover_Type"].value_counts()
```
